# 5. Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
a) Top 10 test teams in men’s cricket along with the records for matches, points and rating.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
pages=requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/test")
```

```python
pages
```

```python
soup5=BeautifulSoup(pages.content,"html.parser")
```

```python
soup5
```

```python
team = soup5.find_all('span',class_="u-hide-phablet")

team_name = []

for i in team:

    team_name.append(i.text)
```

```python
team_name
```

```python
team_matches=[]
```

```python
for i in soup5.find_all('td',class_="rankings-block__banner--matches"):
    team_matches.append(i.text)
for j in soup5.find_all('td',class_="table-body__cell u-center-text"):
        team_matches.append(j.text)
```

```python
team_matches
```

```python
a=team_matches[:2]
b=team_matches[3::2]
e=(a+b)
e
```

```python
team_points=[]
```

```python
for i in soup5.find_all('td',class_="rankings-block__banner--points"):
    team_points.append(i.text)
for j in soup5.find_all('td',class_="table-body__cell u-center-text"):
        team_points.append(j.text)
```

```python
team_points
```

```python
team_points1=team_points[::2]
team_points1
```

```python
team_ratings=[]
```

```python
for i in soup5.find_all('td',class_="rankings-block__banner--rating"):
    team_ratings.append(i.text)
for j in soup5.find_all('td',class_="table-body__cell u-text-right rating"):
        team_ratings.append(j.text)
```

```python
team_ratings
```

```python
data_odi=list(zip(team_name,e,team_points1,team_ratings))
df1=pd.DataFrame(data_odi,columns=["Team_name","Team_matches","Team_points","Team_ratings"])
df1
```
